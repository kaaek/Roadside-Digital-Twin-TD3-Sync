# Result analysis
Load saved metrics and generate paper figures from packaged modules.

```python
import json
from pathlib import Path
import pandas as pd

metrics_path = Path('../results/metrics/monte_carlo_results.json')
if metrics_path.exists():
    results = json.loads(metrics_path.read_text())
    rows = []
    for method, metrics in results.items():
        rows.append({
            'method': method,
            'weighted_aodt_mean': metrics.get('weighted_aodt_mean_float'),
            'weighted_aodt_std': metrics.get('weighted_aodt_std_float'),
            'violations_mean': metrics.get('violation_count_mean_float'),
        })
    display(pd.DataFrame(rows))
else:
    print('Run scripts/run_monte_carlo.py first.')
```
